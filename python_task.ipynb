## Employee Dataset Analysis with Pandas

```python
import pandas as pd

# Create an employee dataset
data = {
    'Employee ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Gita', 'Sarita', 'Chandan', 'Navin', 'Priti', 'Suresh', 'Ram', 'Sara', 'Nishant', 'Jully'],
    'Department': ['HR', 'Engineering', 'Marketing', 'Engineering', 'HR', 'Sales', 'Engineering', 'Marketing', 'Sales', 'HR'],
    'Age': [30, 24, 35, 29, 41, 28, 32, 26, 38, 31],
    'Salary': [60000, 75000, 80000, 78000, 90000, 65000, 82000, 70000, 95000, 62000]
}

employee_df = pd.DataFrame(data)
```

### Employee Dataset

```python
# Display the created DataFrame
display(employee_df.head())
```

```python
# Calculate basic statistics
total_employees = employee_df['Employee ID'].count()
average_salary = employee_df['Salary'].mean()
average_age = employee_df['Age'].mean()

print(f"Total Employees: {total_employees}")
print(f"Average Salary: ${average_salary:,.2f}")
print(f"Average Age: {average_age:.2f} years")
```

```python
# Count employees in each department
employees_by_department = employee_df['Department'].value_counts()

print("\nEmployees by Department:")
display(employees_by_department)
```
